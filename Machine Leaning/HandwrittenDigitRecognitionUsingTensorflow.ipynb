```
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
import math
from keras.datasets import mnist
```

```
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```
X_train_clean = []
for i in range(X_train.shape[0]):
    X_train_clean.append(X_train[i].reshape(784))
X_train = (np.array(X_train_clean)) / 255
```

```
Y_train_clean = []
for i in range(Y_train.shape[0]):
    Y_train_clean.append(np.array([0.0 if Y_train[i] != j else 1.0 for j in range(10)]))
Y_train = np.array(Y_train_clean)
```

```
def predict_using_nn(X_train, Y_train, hidden_units, activations, epochs, learning_rate):
    X = tf.placeholder(tf.float64, shape=X_train.shape)
    Y = tf.placeholder(tf.float64, shape=Y_train.shape)
    W = dict()
    b = dict()
    Z = dict()
    A = dict()
    for i in range(len(hidden_units)+1):
        if i == 0:
          
            # Input layer
            W[str(i+1)] = tf.Variable(tf.random_normal([X_train.shape[1], hidden_units[i]], dtype=tf.float64), name='W'+str(i+1))
            b[str(i+1)] = tf.Variable(tf.random_normal([1, hidden_units[i]], dtype=tf.float64), name='b'+str(i+1))
            Z[str(i+1)] = tf.add(tf.matmul(X_train, W[str(i+1)]), b[str(i+1)])
            
        elif i == len(hidden_units):
                    
            # Output layer
            W[str(i+1)] = tf.Variable(np.random.randn(hidden_units[i-1], Y_train.shape[1]), name='W'+str(i+1))
            b[str(i+1)] = tf.Variable(np.random.randn(1, Y_train.shape[1]), name='b'+str(i+1))
            Z[str(i+1)] = tf.add(tf.matmul( A[str(i)], W[str(i+1)]), b[str(i+1)])
            
        else:
            W[str(i+1)] = tf.Variable(np.random.randn(hidden_units[i-1], hidden_units[i]), name='W'+str(i+1))
            b[str(i+1)] = tf.Variable(np.random.randn(1, hidden_units[i]), name='b'+str(i+1))
            Z[str(i+1)] = tf.add(tf.matmul(A[str(i)], W[str(i+1)]), b[str(i+1)])
            
            
        if activations[i] == 'tanh':
            A[str(i+1)] = tf.nn.tanh(Z[str(i+1)])
        elif activations[i] == 'relu':
            A[str(i+1)] = tf.nn.relu(Z[str(i+1)])
        elif activations[i] == 'sigmoid':
            A[str(i+1)] = tf.nn.sigmoid(Z[str(i+1)])
        else:
            A[str(i+1)] = Z[str(i+1)]
            
    Y_pred = A[str(len(hidden_units) + 1)]
    Y_pred_log = tf.log(Y_pred)
    Y_pred_log_1 = tf.log(1 + 1e-10 - Y_pred)
    
    J = -tf.reduce_mean(tf.reduce_sum(tf.add(tf.multiply(Y, Y_pred_log), tf.multiply(1 - Y, Y_pred_log_1)), 1))
    
    optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(J)

    init = tf.global_variables_initializer()
    sess = tf.Session()
    sess.run(init)
    for i in range(epochs):
        _, J_opt, output = sess.run([optimizer, J, Y_pred_log], feed_dict={X: X_train, Y: Y_train})
        print(i, J_opt)
        if i == epochs - 1:
            return output
```

```
Y_pred = predict_using_nn(np.float64(X_train), np.float64(Y_train), [2500, 2000, 1500, 1000, 500], ['sigmoid', 'sigmoid', 'sigmoid', 'sigmoid', 'sigmoid', 'sigmoid'], 900, 0.01)
```

```
print("Accuracy:", np.sum(np.sum(np.argmax(Y_pred, axis=1) == np.argmax(Y_train, axis=1))) / 60000)
```
